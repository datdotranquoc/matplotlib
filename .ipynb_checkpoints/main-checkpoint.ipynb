```python
x = np.linspace(0,10,1000)
x
```

```python
plt.plot(x, np.sin(x), linestyle = 'dashed');
plt.xlabel("Bien x")
plt.ylabel("Sin(X)")
plt.xlim([0,4])
plt.ylim([-0.7,0.6])
```
