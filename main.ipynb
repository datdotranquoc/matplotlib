<h1>object-oriented API</h1>

```python
#1 Prepare the data
x = [1,2,3,4]
y = [11,22,33,44]

#2 Setup plot
fig, ax = plt.subplots(figsize = (5,5))#Figure size = width and height of the plot
#ax: axes object

#3: plot the data
ax.plot(x,y)

ax.set(title="A simple plot",
       xlabel = "x-axis",
       ylabel = "y-axis");
```

<h1>Most common type of matplotlib plots</h1>
<p>- line
    - scatter

```python
#Line
#create an array
x = np.linspace(0,10,100)
#function of linespace is create a array with std value between start and stop
x[:5]
```

```python
#OO API plot a line chart
fig, ax = plt.subplots()
ax.plot(x, x**3);
```

```python
#scatter
#pylot api
plt.scatter(x, np.exp(x)); #y = e**x, exp: exponential
```

```python
#OO API
#prepare new data:
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

color = rng.rand(100)
sizes = 1000*rng.rand(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x,y, s=sizes, c = color, cmap='viridis', alpha=0.3);

fig.colorbar(img1);
```

```python
#3. Bar(Vertical, Horizontal)
#Prepare the data
soft_drink_prices = {"CocaCola": 15,
                     "Pepsi": 12,
                     "Fanta": 18}
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax.set(title="Bach Hoa Xanh's goods", ylabel="Price $");
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()),list(soft_drink_prices.values()));
#we use list because barh cannot take key value from dict so we have to convert to list
```

```python
#4. Histogram
np.random.seed(42)#seed: to make sure the sequence will be fixed
stu_height = np.random.normal(170, 10, 250)#normal: mean, std, variance
stu_height[:10]
```

```python
#pyplot API
plt.hist(stu_height, bins = 30);#bins: determine and seperate the value
```

```python
#Subplot
#option 1: 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize =(10,5));
#plot data to each axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax4.hist(stu_height, bins = 30);
```

```python
#option 2:
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize =(10,5));
#plot data to each axis
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax[1,1].hist(stu_height, bins = 30);
```
